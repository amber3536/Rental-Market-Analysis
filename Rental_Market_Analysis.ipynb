```python
import pandas as pd
import requests
import json
import time
import csv


# show all columns
pd.set_option('display.max_columns', None)
```

```python
city = 'san marcos'
state = 'ca'
search_str = city + ', ' + state
```

```python
from google.colab import userdata
rapid_api_key = userdata.get('rapid_api_key')
```

```python
# get data
url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"

querystring = {"location":"San Marcos, ca","home_type":"Houses","maxPrice":"900000"}

headers = {
	"X-RapidAPI-Key": rapid_api_key,
	"X-RapidAPI-Host": "zillow-com1.p.rapidapi.com"
}

z_for_sale_resp = requests.request("GET", url, headers=headers, params=querystring)

# transform to json
z_for_sale_resp_json = z_for_sale_resp.json()
z_for_sale_resp_json
```

```python
df_z_for_sale = pd.json_normalize(data=z_for_sale_resp_json['props'])
print('Num of rows:', len(df_z_for_sale))
print('Num of cols:', len(df_z_for_sale.columns))
df_z_for_sale
```

```python
df_homes = df_z_for_sale[['price',  'address','livingArea', 'bathrooms', 'bedrooms']]
df_homes
```

```python
df_homes = df_homes.assign(rent = '0')
```

```python
address = []
size = []

for i, r in df_homes.iterrows():
    address.append(r['address'])
    size.append(r['livingArea'])

length = len(df_homes)
length
```

```python
from google.colab import userdata
rapid_api_key_rental = userdata.get('rapid_api_key_rental')
```

```python
url = "https://realtymole-rental-estimate-v1.p.rapidapi.com/rentalPrice"

for i in range(length):
    querystring = {"address":address[i],"propertyType":"Single Family", "squareFootage":size[i], "compCount":"5"}

    headers = {
        "X-RapidAPI-Key": rapid_api_key_rental,
        "X-RapidAPI-Host": "realtymole-rental-estimate-v1.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)
    response = response.json()

    df_rental = pd.json_normalize(data=response)
    df_homes.at[i,'rent'] = df_rental['rent'].values[0]
    time.sleep(.5)
```

```python
df_homes
```

```python
rentRatio = []
sizeRatio = []

for i, r in df_homes.iterrows():
    rentRatio.append(12* r['rent']/r['price'] *100)
    sizeRatio.append(r['price']/r['livingArea'])
```

```python
df_homes['capRate'] = rentRatio
df_homes['dollarsPerSqft'] = sizeRatio
```
